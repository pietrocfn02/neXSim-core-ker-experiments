```python
SEEDS = ['actor', 
         'amusement_park', 
         'band', 
         'championship', 
         'chemical_element', 
         'company', 
         'comune', 
         'dialect', 
         'film', 
         'scientist', 
         'sculpture', 
         'university', 
         'vehicle', 
         'video_game']


SIZES = [2,3,4,5,6]
```

```python
real_time = None
algorithm_time = None


for seed in SEEDS:
    for size in SIZES:
        _tmp = pd.read_csv(f'../pyrunlim_output_true_true/{seed}/{size}.csv', sep=';')
        _tmp2 = pd.read_csv(f'../24-09-13/{seed}/{size}.csv', sep=';')
        if real_time is None:
            real_time = _tmp
        else:
            real_time = pd.concat([real_time, _tmp])
        if algorithm_time is None:
            algorithm_time = _tmp2
        else:
            algorithm_time = pd.concat([algorithm_time, _tmp2])
        

print(real_time.columns)
print(real_time.shape)

# drop the Unnamed: 20 column
#new_core_real_time = new_core_real_time.drop('Unnamed: 20', axis=1)

print(algorithm_time.columns)
print(algorithm_time.shape)

# drop the Unnamed: 20 column

algorithm_time = algorithm_time.drop('Unnamed: 20', axis=1)


```

```python
# i want the columns: unit, summary_tilde_computation_time and newker_computation_time and a new column in the end with the sum of summary_tilde_computation_time and newker_computation_time


algorithm_time = algorithm_time[['UNIT', 'SUMMARY_TILDE_COMPUTATION_TIME', 'NEWKER_COMPUTATION_TIME']]

algorithm_time['TOTAL_ALGORITHM_TIME'] = algorithm_time['SUMMARY_TILDE_COMPUTATION_TIME'] + algorithm_time['NEWKER_COMPUTATION_TIME']


algorithm_time
```

```python
# i want "unit" "seed" "size" and "realtime" columns (CAPSLOCK)

real_time = real_time[['UNIT', 'SEED', 'SIZE', 'REALTIME']]
real_time.columns = ['UNIT', 'SEED', 'SIZE', 'REALTIME']
real_time
```

```python
# i want to merge the two dataframes on the UNIT column

merged = pd.merge(algorithm_time, real_time, on='UNIT')


merged
```

```python
# i want all in seconds (SUMMARY_TILDE_COMPUTATION_TIME, NEWKER_COMPUTATION_TIME, TOTAL_ALGORITHM_TIME are in milliseconds)

merged['SUMMARY_TILDE_COMPUTATION_TIME'] = merged['SUMMARY_TILDE_COMPUTATION_TIME'] / 1000

merged['NEWKER_COMPUTATION_TIME'] = merged['NEWKER_COMPUTATION_TIME'] / 1000

merged['TOTAL_ALGORITHM_TIME'] = merged['TOTAL_ALGORITHM_TIME'] / 1000


merged
```

```python
# i want a new column "DIFFERENCE" which is TOTAL_ALGORITHM_TIME - REALTIME

merged['DIFFERENCE'] = merged['REALTIME'] - merged['TOTAL_ALGORITHM_TIME']
```

```python
# count the number of rows where the difference is negative

merged[merged['DIFFERENCE'] < 0].shape[0]
```

```python
#show the case where the difference is negative

merged[merged['DIFFERENCE'] < 0]
```

```python
# i want "difference in percentage" which is total_algorithm_time : real_time = x : 100

merged['DIFFERENCE_PERCENTAGE'] = (merged['TOTAL_ALGORITHM_TIME'] * 100) / merged['REALTIME']
```

```python
merged
```

```python
# i want max adn min difference in percentage

print(merged['DIFFERENCE_PERCENTAGE'].max())
print(merged['DIFFERENCE_PERCENTAGE'].min())
print(merged['DIFFERENCE_PERCENTAGE'].mean())
```

```python
# i want the percentiles of the difference in percentage

print(merged['DIFFERENCE_PERCENTAGE'].quantile([0.25, 0.5, 0.75, 0.9, 0.95, 0.99]))
```

```python
# how many cases are above 5% difference

print(merged[merged['DIFFERENCE_PERCENTAGE'] < 25].shape[0])
```
