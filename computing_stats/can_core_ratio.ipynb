```python
import pandas as pd
import numpy as np

import os
```

```python
SEEDS = ['actor', 
         'amusement_park', 
         'band', 
         'championship', 
         'chemical_element', 
         'company', 
         'comune', 
         'dialect', 
         'film', 
         'scientist', 
         'sculpture', 
         'university', 
         'vehicle', 
         'video_game']


SIZES = [2]
```

```python
_can = None

print(os.listdir('../experiments_can_wt'))

for seed in SEEDS:
    for size in SIZES:
        _tmp = pd.read_csv(f'../experiments_can_wt/{seed}/{size}.csv', sep=';')
        if _can is None:
            _can = _tmp
        else:
            _can = pd.concat([_can, _tmp])

print(_can.columns)
print(_can.shape)

# drop the Unnamed: 9 column
_can = _can.drop('Unnamed: 9', axis=1)

#print(_can.columns)
#print(_can.shape)


print(_can.head())
```

```python
# read the names of the subfolders in experiments

print(os.listdir('../24-09-13'))

merged = None


for seed in SEEDS:
    for size in SIZES:
        _tmp = pd.read_csv(f'../24-09-13/{seed}/{size}.csv', sep=';')
        if merged is None:
            merged = _tmp
        else:
            merged = pd.concat([merged, _tmp])

# drop the "Unnamed: N"-like columns

to_remove = [col for col in merged.columns if 'Unnamed' in col]

for col in to_remove:
    merged = merged.drop(col, axis=1)

#merged = merged.drop('Unnamed: 20', axis=1)

print(merged.columns) 
print(merged.shape)

        

```

```python
to_drop = []

# drop columns that are not useful for the analysis

to_drop = [col for col in merged.columns if '_TIME' in col]


# need to drop: OLDKER_SIZE, WT_OLDKER_SIZE, OLDCORE_SIZE, WT_OLCORE_SIZE

to_drop += ['OLDKER_SIZE', 'WT_OLDKER_SIZE', 'OLDCORE_SIZE', 'WT_OLDCORE_SIZE']

# drop the columns


for col in to_drop:
    merged = merged.drop(col, axis=1)
```

```python
# add to "merged" the columns "can_size" and "wt_can_size" joining on "unit"

merged = pd.merge(merged, _can, on=['SEED','UNIT', 'UNIT_SIZE', 'SUMMARY_SIZES', 'WT_SUMMARY_SIZES'], how='inner')

print(merged.columns)

```

```python
merged
```

```python
# drop "UNIT_SIZE", "SUMMARY_SIZES", "WT_SUMMARY_SIZES", "SUMMARY_TILDE_SIZES", "NEWKER_SIZE", "WT_NEWKER_SIZE", "SUMMARY_COMPUTATION_TIME", "CAN_COMPUTATION_TIME"

to_drop = ['UNIT_SIZE', 'SUMMARY_SIZES', 'WT_SUMMARY_SIZES', 'SUMMARY_TILDE_SIZES', 'NEWKER_SIZE', 'WT_NEWKER_SIZE', 'SUMMARY_COMPUTATION_TIME', 'CAN_COMPUTATION_TIME']

for col in to_drop:
    merged = merged.drop(col, axis=1)
```

```python
merged
```

```python
# compute CAN-CORE-ratio and WT-CAN-CORE-ratio

merged['CAN_CORE_RATIO'] = merged['CAN_SIZE'] / merged['NEWCORE_SIZE']
merged['WT_CAN_CORE_RATIO'] = merged['WT_CAN_SIZE'] / merged['NEWCORE_SIZE']

merged
```

```python
# GROUPED BY SEED, AND IN OVERALL, COMPUTE MEAN, MIN, MAX RATIOS

grouped = merged.groupby('SEED').agg({'CAN_CORE_RATIO': ['mean', 'min', 'max'], 'WT_CAN_CORE_RATIO': ['mean', 'min', 'max']})

# ADD THE 'OVERALL' ROW (as another seed)

overall = merged.agg({'CAN_CORE_RATIO': ['mean', 'min', 'max'], 'WT_CAN_CORE_RATIO': ['mean', 'min', 'max']})


grouped

overall

```
