```python
# open all the files in ../esperimenti_con_old_core

INPUT_FILE = "../pyrunlim_first20_output/out.csv"
```

```python
# Dataframe from "2_comune.csv", "2_ce.csv", "2_ap.csv"

df = pd.read_csv(INPUT_FILE, sep=';')
```

```python
df
```

```python
df = df.drop(columns=['Unnamed: 2'])
```

```python
# Count the number of "TIMEOUT" in the "REALTIME" column

df['TIMEOUT'] = df['REALTIME'].apply(lambda x: 1 if x == "TIMEOUT" else 0)

df
```

```python
# Sum the number of "TIMEOUT" in the "TIMEOUT" column
# and report it in a new table with the number of rows and the number of timeouts

_nrows = df['TIMEOUT'].sum()

_ntimeouts = df['TIMEOUT'].count()


print ("Number of rows: ", _nrows)
print ("Number of timeouts: ", _ntimeouts)
print ("Timeout rate: ", _nrows/_ntimeouts*100, "%")
```
