```python
INPUT_FOLDER = '../24-09-13'

SEEDS = ['actor', 
         'amusement_park', 
         'band', 
         'championship', 
         'chemical_element', 
         'company', 
         'comune', 
         'dialect', 
         'film', 
         'scientist', 
         'sculpture', 
         'university', 
         'vehicle', 
         'video_game']


SIZES = [2,3,4,5,6]
```

```python
merged = None
```

```python
for seed in SEEDS:
    for size in SIZES:
        _tmp = pd.read_csv(f'../24-09-13/{seed}/{size}.csv', sep=';')
        if merged is None:
            merged = _tmp
        else:
            merged = pd.concat([merged, _tmp])

# drop the "Unnamed: 15" column
merged = merged.drop('Unnamed: 20', axis=1)

print(merged.columns) 
print(merged.shape)
```

Drop the columns relative to the size and to the "OLD" algorithm

```python
# drop the following columns:
"""
    'SUMMARY_SIZES'
    'WT_SUMMARY_SIZES'
    'SUMMARY_TILDE_SIZES'
    'OLDKER_SIZE'
    'WT_OLDKER_SIZE'
    'NEWKER_SIZE'
    'WT_NEWKER_SIZE'
    'OLDCORE_SIZE'
    'WT_OLDCORE_SIZE'
    'NEWCORE_SIZE'
    'WT_NEWCORE_SIZE'
    'OLDKER_COMPUTATION_TIME'
    'OLDCORE_COMPUTATION_TIME'
"""


merged = merged.drop(['SUMMARY_SIZES',
                          'WT_SUMMARY_SIZES',
                          'SUMMARY_TILDE_SIZES',
                          'OLDKER_SIZE',
                          'WT_OLDKER_SIZE',
                          'NEWKER_SIZE',
                          'WT_NEWKER_SIZE',
                          'OLDCORE_SIZE',
                          'WT_OLDCORE_SIZE',
                          'NEWCORE_SIZE',
                          'WT_NEWCORE_SIZE',
                          'OLDKER_COMPUTATION_TIME',
                          'OLDCORE_COMPUTATION_TIME'], axis=1)
```

```python
# i want two different datasets:
# 1. one with 'ker' (summary_tilde + 'ker')
# 2. one with 'core' (summary + 'core')

ker = merged[['SEED', 'UNIT', 'UNIT_SIZE', 'SUMMARY_TILDE_COMPUTATION_TIME', 'NEWKER_COMPUTATION_TIME']]

core = merged[['SEED', 'UNIT', 'UNIT_SIZE', 'SUMMARY_COMPUTATION_TIME', 'NEWCORE_COMPUTATION_TIME']]
```

```python
# in ker, total computation time is the sum of 'SUMMARY_TILDE_COMPUTATION_TIME' and 'NEWKER_COMPUTATION_TIME'

ker['TOTAL_COMPUTATION_TIME'] = ker['SUMMARY_TILDE_COMPUTATION_TIME'] + ker['NEWKER_COMPUTATION_TIME']
```

```python

core['TOTAL_COMPUTATION_TIME'] = core['SUMMARY_COMPUTATION_TIME'] + core['NEWCORE_COMPUTATION_TIME']
```

```python
def max_comp_time(df, column):

    table = pd.DataFrame(columns=['SEED'] + SIZES + ['OVERALL'])

    for seed in SEEDS:
        row = {'SEED': seed}
        for size in SIZES:
            val = df[(merged['SEED'] == seed) & (merged['UNIT_SIZE'] == size)][f'{column}'].max()
            row[size] = val
        val = df[df['SEED'] == seed][f'{column}'].max()
        row['OVERALL'] = val

        table = pd.concat([table, pd.DataFrame([row])])

    return table


def to_seconds(df):
    for _col in df.columns:
        if _col != 'SEED':
            df[_col] = df[_col] / 1000
            df[_col] = df[_col].round(5)

    return df


```

```python
table4 = max_comp_time(core, 'NEWCORE_COMPUTATION_TIME')
table5 = max_comp_time(core, 'SUMMARY_COMPUTATION_TIME')
table6 = max_comp_time(core, 'TOTAL_COMPUTATION_TIME')



table4 = to_seconds(table4)
table5 = to_seconds(table5)
table6 = to_seconds(table6)

table4.to_csv('tables/max_newcore_computation_time.csv', index=False)
table5.to_csv('tables/max_summary_computation_time.csv', index=False)
table6.to_csv('tables/max_summary_newcore_computation_time.csv', index=False)

```
